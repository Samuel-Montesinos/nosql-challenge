5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# finding what needs changed
pipeline = [
    {
        '$addFields': {
            'geocode.latitude': {'$toDouble': '$geocode.latitude'},
            'geocode.longitude': {'$toDouble': '$geocode.longitude'}
        }
    },
    {
        '$out': 'establishments'  # Specify the collection name to overwrite
    }
]

# Execute the aggregation pipeline
collection.aggregate(pipeline)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Setting rating value to integer using pipelines as update_many wasn't working for me
pipeline = [
    {
        '$addFields': {
            'RatingValue': {'$toInt': '$RatingValue'}
        }
    },
    {
        '$out': 'establishments'  # Specify the collection name to overwrite
    }
]

# Execute the aggregation pipeline
collection.aggregate(pipeline)
```

```python
# Check that the coordinates and rating value are now numbers
# verifying no strings in columns
filter = {
    '$or': [
        {'geocode.latitude': {'$type': 'string'}},
        {'geocode.longitude': {'$type': 'string'}},
        {'RatingValue': {'$type': 'string'}}
    ]
}
updated_documents = collection.find(filter)

# Print the updated documents
updated_documents_list = list(updated_documents)
count = len(updated_documents_list)

if count > 0:
    for document in updated_documents:
        print('Strings still in columns')
else:
    print('All strings converted to numbers.')
```
